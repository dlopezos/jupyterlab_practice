```python
def compute_mean(values):
    mean = sum(values) / len(values)
    print (mean)
compute_mean([2, 4, 6, 8])
```

The code above defines a function that computes the mean of the values in a list and prints the results.
Below are two mathematical expressions for the mean.

$
\bar{x} = \frac{x_1 + x_2 + \cdots + x_n}{n}
\$

or

$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
\$
